```python
import yfinance as yf
import pandas as pd

# Define tickers and date range
tickers = ['TSLA', 'BND', 'SPY']
start_date = '2015-01-01'
end_date = '2025-01-31'

# Download historical data
data = yf.download(tickers, start=start_date, end=end_date)

# Save to CSV for reference
data.to_csv("historical_data.csv")

print("Data loaded successfully!")
```

```python
# Check for missing values
print(data.isnull().sum())

# Fill missing values (choose a method that fits your data best)
data.fillna(method='ffill', inplace=True)  # Forward-fill
data.fillna(method='bfill', inplace=True)  # Backward-fill

# Check basic statistics
print(data.describe())

# Ensure correct data types
print(data.dtypes)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize only numerical columns (e.g., Open, High, Low, Close, Volume)
scaler = MinMaxScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns, index=data.index)

print(data_scaled.head())
```

```python
import matplotlib.pyplot as plt

# Plot Closing Prices for TSLA
data[('Close', 'TSLA')].plot(figsize=(10, 6), title="TSLA Closing Prices")
plt.ylabel("Price")
plt.show()

# Calculate daily returns for TSLA
daily_returns = data[('Close', 'TSLA')].pct_change()
daily_returns.plot(figsize=(10, 6), title="TSLA Daily Returns")
plt.ylabel("Percentage Change")
plt.show()

# Rolling statistics for volatility analysis
rolling_mean = data[('Close', 'TSLA')].rolling(window=20).mean()  # 20-day average
rolling_std = data[('Close', 'TSLA')].rolling(window=20).std()    # 20-day standard deviation

# Plot rolling statistics
plt.figure(figsize=(10, 6))
plt.plot(data[('Close', 'TSLA')], label="Close")
plt.plot(rolling_mean, label="Rolling Mean (20 days)", linestyle='--')
plt.plot(rolling_std, label="Rolling Std Dev (20 days)", linestyle='--')
plt.legend()
plt.title("Rolling Mean and Standard Deviation for TSLA")
plt.show()
```

```python
# Identify outliers using standard deviation
mean = daily_returns.mean()
std_dev = daily_returns.std()
outliers = daily_returns[(daily_returns > mean + 3*std_dev) | (daily_returns < mean - 3*std_dev)]

# Print outliers
print("Outliers:")
print(outliers)
```

```python
import matplotlib.pyplot as plt

# Plot Close prices
data['Close_TSLA'].plot(figsize=(12, 6), title="TSLA Close Prices with Outliers", color='blue')

# Mark outliers
outlier_dates = daily_returns[daily_returns.abs() > daily_returns.std() * 3].index  # Dates of outliers
plt.scatter(outlier_dates, data['Close_TSLA'].loc[outlier_dates], color='red', label='Outliers')

plt.legend()
plt.ylabel("Price")
plt.xlabel("Date")
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Extract Tesla's Close price
tsla_close = data['Close_TSLA']

# Ensure the series does not contain missing values
tsla_close = tsla_close.dropna()

# Decompose the time series
decomposition = seasonal_decompose(tsla_close, model='additive', period=365)

# Plot the decomposition
decomposition.plot()
plt.suptitle("Seasonal Decomposition of TSLA Close Prices", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the rolling mean and rolling standard deviation
rolling_mean = data['Close_TSLA'].rolling(window=20).mean()  # 20-day rolling mean
rolling_std = data['Close_TSLA'].rolling(window=20).std()    # 20-day rolling standard deviation

# Plot the original closing prices, rolling mean, and rolling standard deviation
plt.figure(figsize=(12, 6))

# Original closing prices
plt.plot(data['Close_TSLA'], label="TSLA Close Prices", color='blue', alpha=0.6)

# Rolling mean
plt.plot(rolling_mean, label="20-Day Rolling Mean", color='orange', linestyle='--')

# Rolling standard deviation (volatility)
plt.plot(rolling_std, label="20-Day Rolling Std Dev (Volatility)", color='red', linestyle='--')

# Chart customization
plt.title("Tesla (TSLA) Rolling Mean and Volatility")
plt.xlabel("Date")
plt.ylabel("Price / Volatility")
plt.legend()
plt.show()
```
